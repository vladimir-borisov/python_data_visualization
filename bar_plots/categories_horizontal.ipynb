```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# generate sample data
values = np.arange(50, 0, -5)
products = np.arange(65, 75, 1)
products = list(map(chr, products))
```

```python
sns.set(style = "ticks")

# set figure size and resolution (dpi)
plt.figure(figsize = (10, 7), dpi = 150)

# horizontal bar plot from seaborn
ax = sns.barplot(x = values, y = products, color = "navy")

# hide top and right axis
sns.despine()

# turn on vertical axis lines
ax.xaxis.grid(True)
```
