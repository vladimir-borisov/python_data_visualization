```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = []

data.extend(['One' for i in range(10)])
data.extend(['Two' for i in range(7)])
data.extend(['Three' for i in range(9)])
data.extend(['Four' for i in range(4)])
data.extend(['Five' for i in range(6)])
```

```python
sns.set(style="ticks", rc={"lines.linewidth": 5, 'axes.grid' : False})

# set figure size and resolution (dpi)
plt.figure(figsize = (10, 7), dpi = 150)

# count plot just calculate count of each element in x 
# so x values it is all unique values in data and y values is how
# many time each value appears in data
ax = sns.countplot(x = data, palette = sns.light_palette("navy", reverse = True))

#create labels for axis
ax.set_ylabel('', fontsize = 22)
ax.set_xlabel('', fontsize = 22)

# make tick's labes along both X and Y axis bigger 
ax.tick_params(labelsize = 13)

# turn off top and right axis
sns.despine()
```
