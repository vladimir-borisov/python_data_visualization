```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
```

```python
# generate data for the plot just 2000 values with normal (gaussian) distribution
center = 0
deviation = 5
data = np.random.normal(loc = 0, scale = deviation, size = 2000) 
```

```python
# set seaborn style and a few parameters 
sns.set(style = "ticks", rc = {"lines.linewidth": 3, 'axes.grid' : False})

# set plot size
plt.figure(figsize = (10, 7))

# define bars positions. 
# NOTE: these are left parts of the bins not the center of the bins
bins = [i for i in range(-20, 20, 1)]

# create seaborn distplot
# read more: https://seaborn.pydata.org/generated/seaborn.distplot.html#seaborn.distplot
ax = sns.distplot(data, bins = bins, color = '#112ca6', norm_hist = False)

#create labels and limits for axis
ax.set_ylabel('', fontsize = 20)
ax.set_xlabel('', fontsize = 20, fontname = 'Georgia')

# chosen by hands just for nice view
ax.set_xlim((-20, 20))

# we use a hack for Y scale by default in seaborn it is density 
# but we want to see % of each bar value in all data, so 1 = 100% 
# NOTE: 100% = bin_length * bin_height(Y value), so if tou have bin_length = 0.5 then 1 point on Y = 50% !!!
# 0.08 = 8% we chose it for good view, try to change it to (0, 1) and understand the idea
ax.set_ylim((0, 0.08))

# it's better to see 8% instead of 0.08 so for each value on the Y axis we used x * 100
# try to comment next line and see what will be :)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(int(x * 100)) + '%'))

# we wanna have 5 points between numbers on X axis, it looks nice
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))

# hide the right and top axis we don't need them 
sns.despine()
```
