```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import seaborn as sns
```

```python
data1 = np.random.normal(-10, 5, 1000)
data2 = np.random.normal(10, 5, 1000)
```

```python
sns.set(style="ticks", rc={"lines.linewidth": 5, 'axes.grid' : False})

# set figure size and resolution (dpi)
plt.figure(figsize = (12, 7), dpi = 150)

# add 2 distribution plots on the same axes
ax = sns.distplot(data1, color = '#02126C', hist = False, label = '', kde_kws={"shade": True},)
ax = sns.distplot(data2, color = '#ff0000', hist = False, label = '', kde_kws={"shade": True},)

# change tick label size 
ax.tick_params(labelsize = 17)

# hide the top and right axis lines
sns.despine()

#create labels for axis
ax.set_ylabel('', fontsize = 22)
ax.set_xlabel('Values', fontsize = 22)


# if you wanna see % values on Y axis instead of coefficients like 0.01
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: str(int(x * 100)) + '%'))
```
