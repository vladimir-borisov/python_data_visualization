```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import random
#you can see what is tick from the 'matplotlib_figure_structure.png' in the root of the repository
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
```

```python
# generate dates for some period of time, only dates, no exact time needed 
dates = pd.date_range(start = '2020-03-01', end = '2020-05-05', freq = 'D')

# we want pure python date types
dates = dates.to_pydatetime()

count_of_dates = len(dates)

# it is a base line where all values are equal 50
values1 = [50 for i in range(count_of_dates)]

# for each element we take 50 and + or - a random value [0, 1) * 5
values2 = [50 + (random.random() - 0.5) * 5  for i in range(count_of_dates)]
```

```python
#set the seaborn settings
sns.set(style="ticks", rc={"lines.linewidth": 5, 'axes.grid' : False})

# set figure size and x ticks rotation and plot title
plt.figure(figsize = (21, 7), dpi = 150) # dpi is a resolution measure of matplotlib figure
plt.xticks(rotation = 30)
plt.title('Timeseries plot', fontsize = 25, fontname = 'Georgia')


# create seaborn line as matplotlib axes 
ax = sns.lineplot(x = dates, y = values1, color = '#038cfc', label = 'True values')
ax = sns.lineplot(x = dates, y = values2, color = '#2800c7', label = 'Predicted values')
# NOTE: these two plots will be on the same axes, so we use 'ax' for both

#Create readable dates format
form = mdates.DateFormatter('%Y-%m-%d')
# You can pass you own function to the FuncFormatter
ax.xaxis.set_major_formatter(ticker.FuncFormatter(form))

#count of days between two dates on the X axis
ax.xaxis.set_major_locator(ticker.MultipleLocator(7))

# for the best look we limit Y shown values
ax.set_ylim((40, 60))

# plot vertical grid lines. It helps us to match value-date pairs
ax.xaxis.grid(True)

# try to set 0.05 and you will see why
ax.margins(x = 0)

# make tick's labes along both X and Y axis bigger 
ax.tick_params(labelsize = 15)

#turn off up and right plot borders
sns.despine()

# change legend(table with labels) size 
plt.legend(prop = {'size': 15})
```
