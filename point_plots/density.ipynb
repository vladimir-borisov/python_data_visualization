```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
```

```python
# data preparation for the plot

# create 1000 randomly distributed values 
# function generate values in [0, 1), but we want [0, 100)
x = np.random.rand(1000) * 100

# we want x correlate with y so we take x and add a small random value to each x value
# it is like: y = x + small_random_value
y = x + np.random.rand(1000)
```

```python
joint_kws = {'shade_lowest':False}

g = sns.jointplot(x = x, y = y, kind="kde", joint_kws = joint_kws, height = 8)

# you can use it with pandas dataframe like:
#g = sns.jointplot(x = "value1", y = "value2", data = pandas_dataframe, kind="kde", color = 'b', joint_kws = joint_kws, height = 8)

g.ax_joint.tick_params(labelsize = 13)

g.ax_joint.set_xlim((-15, 115))
g.ax_joint.set_ylim((-15, 115))

g.ax_joint.set_xlabel(xlabel = 'Value 1', fontsize = 18)
g.ax_joint.set_ylabel(ylabel = 'Value 2', fontsize = 18)

sns.despine()
```
