```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#you can see what is tick from the 'matplotlib_figure_structure.png' in the root of the repository
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('../data/timeseries_sample.csv')

#convert object -> datetime64
df['timestamp'] = pd.to_datetime(df['timestamp'])

#get only year, month, day from timestamp
#it is not really needed just for sample
#df['date'] = pd.to_datetime(df['timestamp']).dt.date
```

```python
#set the seaborn settings
sns.set(style="whitegrid", rc={"lines.linewidth": 3})

#some settings for matplotlib plot
plt.figure(figsize = (21, 7))
plt.xticks(rotation = 30)
plt.title('Plot title', fontsize = 25, fontname = 'Georgia')

#create seaborn line as matplotlib axes
ax = sns.lineplot(x = "timestamp", y = "value", data = df, label = 'Line: A')

#sample of usage array for parameter
values = df['value'] + 4
ax2 = sns.lineplot(x = "timestamp", y = values,  data = df, label = 'Line: B')

#Axes styles. Here we use only 'ax' because 'ax' and 'ax2' at the same place on the figure
#Create readable dates format
form = mdates.DateFormatter('%Y-%m-%d')
ax.xaxis.set_major_formatter(ticker.FuncFormatter(form))

#count of days between two date on X - axis
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))

#create labels for axis
ax.set_xlabel('X label', fontsize = 22)
ax.set_ylabel('Y label', fontsize = 22)


#if you don't want auto rescale
#ax.set_ylim((0, 100))
```
